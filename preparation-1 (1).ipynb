```python
import pandas as pd
import os 
import shutil
```

```python
df = pd.read_csv("../input/hostcompetition/test/test.csv", encoding = 'latin1')
```

```python
TARGET_DIR = "./0"
IMAGES_PATH = "../input/hostcompetition/test/DSLR/"   # Do the same for the test label as well. Change path only the train CSV file. 
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Normal Folder Created")
```

```python
cnt = 0

for(i,row) in df.iterrows():
    if row["label"]== 0 and row["modality"]=="DSLR":
        filename=row["id"]
        print("x")
        image_path = os.path.join(IMAGES_PATH,filename+'.png')
        image_copy_path = os.path.join(TARGET_DIR,filename+'.png')
        shutil.copy2(image_path,image_copy_path)
        print("moving image", cnt)
        cnt+=1
```
